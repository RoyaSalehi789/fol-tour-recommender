```python
prolog = Prolog()

prolog.retractall("country(_,_)")
prolog.retractall("region(_,_)")
prolog.retractall("climate(_,_)")
prolog.retractall("budget(_,_)")
prolog.retractall("activity(_,_)")
prolog.retractall("demographics(_,_)")
prolog.retractall("duration(_,_)")
prolog.retractall("cuisine(_,_)")
prolog.retractall("history(_,_)")
prolog.retractall("natural_wonder(_,_)")
prolog.retractall("accommodation(_,_)")
prolog.retractall("language(_,_)")
```

```python
def dfToStr (df, interpretation):
    prefix =  interpretation + "("
    suffix = ")"
    final_str_list = []
    for i in range (df.shape[0]):
        a = []
        a = df.loc[i].tolist()
        
        if "," in a[1]:
            splited_list = a[1].split(", ")
            a[0] = chr(34) + a[0] + chr(34)
            for j in splited_list:
                row_values_str = a[0].lower().replace(" ", "_") + ', ' + j.lower().replace(" ", "_")
                finalStr = prefix + row_values_str + suffix
                final_str_list.append(finalStr)
            
        else:
            a[0] = chr(34) + a[0] + chr(34)
            row_values_str = a[0].lower().replace(" ", "_") + ', ' + a[1].lower().replace(" ", "_")
            finalStr = prefix + row_values_str + suffix
            final_str_list.append(finalStr)
        
    return final_str_list
```

```python
def add_kb(syntax_list):
    for i in range(len(syntax_list)):
        prolog.assertz(syntax_list[i])
```

```python
def create_dict(key_label,value_list):
    value_list = [word.lower().replace(" ", "_") for word in value_list]
    keyword_dict.update({key_label: value_list})
```

```python
def extract_keywords(text, keyword_dict):
    text = re.sub(r'[^a-zA-Z_ ]', '', text)
    words = text.split()
    print(words)

    extracted_keywords = []

    for word in words:
        for key, value in keyword_dict.items():
            if word in value:
                extracted_keywords.append(word)

    return extracted_keywords

user_text = "I would like to travel somewhere in south_america activity with low."

result_keywords = extract_keywords(user_text, keyword_dict)

print(result_keywords)
```

```python
def send_query(extracted_keywords):
    i = 0
    results = []
    for word in extracted_keywords:
        for key, value in keyword_dict.items():
            if word in value:
                if i == 0:
                    query = f"{key}(City, {word})"
                    temp = list(prolog.query(query))
                    results = temp
                else:
                    query = f"{key}(City, {word})"
                    results = list(prolog.query(query))
        temp = [value for value in results if value in temp]
        i += 1
    for result in temp:
        print(result["City"])
    
    
```
