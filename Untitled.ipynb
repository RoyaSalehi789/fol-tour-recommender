```python
def dfToStr (df, interpretation):
    prefix =  interpretation + "("
    suffix = ")"
    final_str_list = []
    for i in range (df.shape[0]):
        a = []
        a = df.loc[i].tolist()
        
        if "," in a[1]:
            splited_list = a[1].split(", ")
            a[0] = chr(39) + a[0] + chr(39)
            for j in splited_list:
                row_values_str = a[0].lower().replace(" ", "_") + ', ' + j.lower().replace(" ", "_")
                row_values_str = row_values_str.replace(".", "")
                row_values_str = row_values_str.replace("'", "")
                finalStr = prefix + row_values_str + suffix
                final_str_list.append(finalStr)
            
        else:
            a[0] = chr(39) + a[0] + chr(39)
            row_values_str = a[0].lower().replace(" ", "_") + ', ' + a[1].lower().replace(" ", "_")
            row_values_str = row_values_str.replace(".", "")
            row_values_str = row_values_str.replace("'", "")
            finalStr = prefix + row_values_str + suffix
            final_str_list.append(finalStr)
        
    return final_str_list
```

```python
def extract_keywords(text, keyword_dict):
    text = re.sub(r'[^a-zA-Z_ ]', '', text).lower()
    words = text.split()

    extracted_keywords = []

    for word in words:
        for key, value in keyword_dict.items():
            if word in value:
                extracted_keywords.append(word)

    return extracted_keywords

user_text = "Embarking on a medium-length journey, I aim to immerse myself in the cultural depth of ancient European cities. With a mid-range budget, my focus will be on experiencing the rich tapestry of history and art that defines Europe's storied past. This adventure will blend the allure of ancient heritage with the comfort of mid-range accommodations, creating a balanced and enriching cultural exploration."

result_keywords = extract_keywords(user_text, keyword_dict)

print(result_keywords)
```

```python
def second_connection_city(connections_dict):
    second_connections_dict = {}
    for key, value in connections_dict.items():
        second_connections_dict[key] = {}
        for i in range(len(value)):
            second_neighbors = connections_dict[connections_dict[key][i]]
            result = [s for s in second_neighbors if s not in value]
            result.remove(key)
                
            second_connections_dict[key][connections_dict[key][i]] = result


    return second_connections_dict
            
                    
```
