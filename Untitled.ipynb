```python
def dfToStr (df, interpretation):
    prefix =  interpretation + "("
    suffix = ")"
    final_str_list = []
    for i in range (df.shape[0]):
        a = []
        a = df.loc[i].tolist()
        
        if "," in a[1]:
            splited_list = a[1].split(", ")
            a[0] = chr(39) + a[0] + chr(39)
            for j in splited_list:
                row_values_str = a[0].lower().replace(" ", "_") + ', ' + j.lower().replace(" ", "_")
                row_values_str = row_values_str.replace(".", "")
                row_values_str = row_values_str.replace("'", "")
                finalStr = prefix + row_values_str + suffix
                final_str_list.append(finalStr)
            
        else:
            a[0] = chr(39) + a[0] + chr(39)
            row_values_str = a[0].lower().replace(" ", "_") + ', ' + a[1].lower().replace(" ", "_")
            row_values_str = row_values_str.replace(".", "")
            row_values_str = row_values_str.replace("'", "")
            finalStr = prefix + row_values_str + suffix
            final_str_list.append(finalStr)
        
    return final_str_list
```

```python
def extract_keywords(text, keyword_dict):
    text = re.sub(r'[^a-zA-Z_ ]', '', text).lower()
    words = text.split()

    extracted_keywords = []

    for word in words:
        for key, value in keyword_dict.items():
            if word in value:
                extracted_keywords.append(word)

    return extracted_keywords

user_text = "As a solo traveler with a passion for ancient cultures, I'm eager to immerse myself in the rich heritage of the Persian language. My journey is not just about sightseeing but delving deep into the linguistic roots of one of the world's oldest languages. Traveling alone will allow me the freedom to explore at my own pace, meet local speakers, and gain a firsthand understanding of the nuances of Persian in its native setting."

result_keywords = extract_keywords(user_text, keyword_dict)

print(result_keywords)
```

```python
def one_way(connected_cities, city):
    prefix =  "directly_connected" + "("
    suffix = ")"
    final_str_list = []
    for i in range (len(connected_cities)):
        row_values_str =city.lower().replace(" ", "_") + ', ' + connected_cities[i].lower().replace(" ", "_")
        row_values_str = row_values_str.replace(".", "")
        row_values_str = row_values_str.replace("'", "")
        finalStr = prefix + row_values_str + suffix
        final_str_list.append(finalStr)
    return final_str_list
```
